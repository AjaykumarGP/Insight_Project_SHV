```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
import numpy as np
```

```python
df = pd.read_csv("highest_cleaned")
```

```python
df.head()
```

```python
my_list = list(df.Corrected_content)
ref_words = [word_tokenize(i) for i in my_list]
len(ref_words)
```

```python
words_tagged = [nltk.pos_tag(i) for i in ref_words]
```

```python
len(words_tagged)
```

```python
from nltk.corpus import wordnet
```

```python
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None
```

```python
words_tagged[0]
```

```python
verb_list=[]
for i in range(len(words_tagged)):
    #verb_list=[]
    for (word, tag) in words_tagged[i]:
        wntag = get_wordnet_pos(tag)
        if wntag is wordnet.VERB:
            verb_list.append(word)
```

```python
noun_list=[]
for i in range(len(words_tagged)):
    #verb_list=[]
    for (word, tag) in words_tagged[i]:
        wntag = get_wordnet_pos(tag)
        if wntag is wordnet.NOUN:
            noun_list.append(word)
```
